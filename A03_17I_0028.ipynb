## **Gene**

```
#Assuming each lecture is of a 3-credit hour course, we need to make sure that each lecture has no more and no less than slots covering 3 hrs worth of lecture
#For this, we need to check for the duration of the lecture, given in minutes. A 60min lecture should have 3 slos per week

class Gene:
  def __init__(self, lect, allocatedRoom, day, startTime):
    self.lect = lect
    self.allocatedRoom = allocatedRoom
    self.day = day
    self.startTime = startTime
  #for printing
  def __repr__(self):
    return "\n\nLecture Details:\n %s \nAllocated Room: %s \nDay: %s \nStarting Time: %s" % (self.lect, self.allocatedRoom, self.day, self.startTime)  
  def __str__(self):
    return "\n\nLecture Details:\n %s \nAllocated Room: %s \nDay: %s \nStarting Time: %s" % (self.lect, self.allocatedRoom, self.day, self.startTime)  
#Gene(Lectures[4], 'room', 'today', '670')
#Lectures[4].allowedClassrooms[2]
```

## **Chromosome**

```
#Generating a Chromosome - a complete solution

def generateChromosomes():
  #Chromosome is a list of genes
  Chromosome = []

  #shuffle Lecture list
  temp = random.sample(Lectures, len(Lectures))

  #allocate a slot, room and day to each lecture
  for x in temp:
    #duration can be 60 minutes, 90 minutes or 180 minutes
    #3 slots needed if 60, 2 if 90, 1 if 180

    if x.duration=='3':
      #startingtime!=after 2pm, only one slot needed
      d = random.choice(weekdays)
      if d=='Friday':
        while 1:
          stime=random.choice(startingTimes[:-4])
          if stime not in ['10.30','11.00','11.30','12.00','12.30','1.00','1.30']:
            Chromosome.append(Gene(x, random.choice(x.allowedClassrooms), d, stime))
            break
      else:
        Chromosome.append(Gene(x, random.choice(x.allowedClassrooms), d, random.choice(startingTimes[:-4])))

    elif x.duration=='1.5':
      #startingtime!=after 3.30pm, two slots needed
      for i in range (0,2):
        d = random.choice(weekdays)
        if d=='Friday':
          while 1:
            stime=random.choice(startingTimes[:-1])
            if stime not in ['12.00','12.30','1.00','1.30']:
              Chromosome.append(Gene(x, random.choice(x.allowedClassrooms), d, stime))
              break
        else:
          Chromosome.append(Gene(x, random.choice(x.allowedClassrooms), d, random.choice(startingTimes[:-1])))

    elif x.duration=='1':
      #startingtime!=after 4pm, three slots needed
      for i in range (0,3):
        d = random.choice(weekdays)
        if d=='Friday':
          while 1:
            stime=random.choice(startingTimes)
            if stime not in ['12.30','1.00','1.30']:
              Chromosome.append(Gene(x, random.choice(x.allowedClassrooms), d, stime))
              break
        else:
          Chromosome.append(Gene(x, random.choice(x.allowedClassrooms), d, random.choice(startingTimes)))

  return Chromosome

#generateChromosomes()
```

## **Crossover**

```
# crossover function takes in two chromosomes and swaps some genes for them

def crossover(A, B):
  #print ("A is: ", len(A))
  #print (A[0])
  #print ("\nB is: ", len(B))
  #print(B[53])

  #length of chromosome A and B is the number of genes in each
  #halving the genes so that we can swap the halves between the 2 chromosomes
  
  offS1=[]
  offS2=[]
  x = int(0.5*(len(A)))
  y = range(0, x)
  for i in y:
    offS1.append(A[i])
    offS2.append(B[i])

  y = range(x,len(A))
  for i in y:
    offS1.append(B[i])
    offS2.append(A[i])

  #this array would store the 4 resultant chromosomes, 2 parents, 2 offsprings
  resultants=[]
  resultants.append(A)
  resultants.append(B)
  resultants.append(offS1)
  resultants.append(offS2)
  return resultants
  
#crossover(Population[0], Population[1])
```

## **Mutation**

```
# mutation function takes in a chromosome and mutates its genes

def mutate(chromosome):
  mutationRate = 0.3
  #generatte a random no for each gene, mutate the genes which have the number smaller than mutation rate
  for gene in chromosome:
    x = random.random()
    if x<mutationRate:
      #print ("Mutating because ", x)
      #y = chromosome.index(gene)
      # mutate gene
      # keep lecture fixed to ensure that cred hrs constraint isnt violated
      sth=gene.allocatedRoom
      while (sth==gene.allocatedRoom):
        sth = random.choice(gene.lect.allowedClassrooms)
      gene.allocatedRoom=sth

      sth=gene.day
      while (sth==gene.day):
        sth = random.choice(weekdays)
      gene.day=sth

      sth=gene.startTime
      while (sth==gene.startTime):
        sth = random.choice(startingTimes)
      gene.startTime=sth
```

## **Roulette Selection Method**
## for Selection of Chromosomes for Crossover

```
# to select which chromosomes to crossover
# Roulette Selection Method is used twice to randomly pick two chromosomes to crossover

def crossoverSelection(pop):
  sumFit = 0
  for chr in pop:
    sumFit+=checkFitness(chr)
  #print("sum is ",sumFit)
  #max = sum([checkFitness(chromosome) for chromosome in population])
  pick = random.uniform(0, sumFit)
  current = 0
  
  for chr in pop:
    current += checkFitness(chr)
    if current > pick:
      return (pop.index(chr))

#generate two random chromosomes to crossover
#parent1 = (crossoverSelection(Population))
#parent2 = (crossoverSelection(Population))
#print ("crossing over chromosome at index", parent1, " and ", parent2)
#crossover(Population[parent1], Population[parent2])
```

## Selecting Chromosomes for Mutation

```
# to select which chromosomes to mutate

def mutateChromosomes(pop):
  mr = 0.1
  for chr in pop:
    x = random.random()
    #print (x)
    if x<mr:
      #print ("Mutating because ", x)
      #y = pop.index(chr)
      #print(y)
      # mutate chromosome
      mutate(chr)

mutateChromosomes(Population)
```

## Function to check if solution exists

```
def chckSol(pop):
  for chr in pop:
      if checkFitness(chr)==70:
        print ("Chromosome at index ", pop.index(chr), " in testPopulation array has max possible fitness. It is the solution.")
        return chr

if chckSol(Population)==None:
  print("No solution yet")
```

# **Main** **Function**

```
# Main Function

def generateSchedule():
  # Step 1 - Create a Population
  testPopulation = []
  for i in range(0,11):
    testPopulation.append(generateChromosomes())
  
  gen =0
  # Step 2 - Check if max fitness 70 is achieved for any of the chromosomes
  sol = chckSol(testPopulation)
  if sol!=None:
    #print(sol)
    return sol
  else:
    print("Generation:", gen)
    print("Max fitness in %: ", (max(checkFitness(c) for c in testPopulation)/70)*100)
  

  while(True):
    # Step 3 - If solution is not found, mutate and crossover

    # Crossover
    par1 = (crossoverSelection(testPopulation))
    par2 = (crossoverSelection(testPopulation))
    resArray = crossover(testPopulation[par1], testPopulation[par2])
    #pick most fit among these 4
    fitArray = []
    fitArray.append(checkFitness(resArray[0]))
    fitArray.append(checkFitness(resArray[1]))
    fitArray.append(checkFitness(resArray[2]))
    fitArray.append(checkFitness(resArray[3]))
    newArray=[]
    for element in fitArray:
      newArray.append(element)
    newArray.sort()

    testPopulation[par1]=resArray[fitArray.index(newArray[3])]
    testPopulation[par2]=resArray[fitArray.index(newArray[2])]

    gen+=0.5

    sol = chckSol(testPopulation)
    if sol!=None:
      print(sol)
      return sol
    else:
      print("Generation:", gen)
      print("Max fitness in %: ", (max(checkFitness(c) for c in testPopulation)/70)*100)

    sad=((max(checkFitness(c) for c in testPopulation))/70)*100
    if gen>55 and sad>85:
      print (sad, "% constraints satisfied")
      return

    # mutate
    mutateChromosomes(testPopulation)
    gen+=0.5
    
    sol = chckSol(testPopulation)
    if sol!=None:
      print(sol)
      return sol
    else:
      print("Generation:", gen)
      print("Max fitness in %: ", (max(checkFitness(c) for c in testPopulation)/70)*100)
  
    sad=((max(checkFitness(c) for c in testPopulation))/70)*100
    if gen>55 and sad>85:
      print (sad, "% constraints satisfied")
      return
    


generateSchedule()
```
